## Data Exploring

### How many rows and how many columns?

```python
num_rows = udemy_df.shape[0]
num_cols = udemy_df.shape[1]
print(f"Your data has {num_rows} rows and {num_cols} columns")
```

### What is the meaning of each row?

Each row represents a specific course with details, providing information about the course's characteristics, ratings, prices, and other relevant information.

### Are there duplicated rows?

```python
num_duplicated_rows = udemy_df.duplicated().sum()
if num_duplicated_rows:
    print(f"Your data has {num_duplicated_rows} duplicated rows")
else:
    print("Your data has no duplicated row!")
```
