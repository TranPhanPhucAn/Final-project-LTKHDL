## With each categorical column, how are values distributed?

```python
udemy_df = udemy_df.drop(columns=['discount_price__price_string','price_detail__price_string'])
```

```python
cat_col_info_df = udemy_df.select_dtypes(include=['object','bool']).drop(columns = ['created','published_time'])
def missing_ratio(s):
    temp=s.isnull().sum()
    result=float(temp)/len(s.index) *100
    return result
def num_values(s):
    temp=s.copy()
    a=list(set(temp))
    has_missing_values = temp.isnull()
    if True in has_missing_values.values:
        return len(a)-1            
    return len(a)
def value_ratios(s):
    dict={}
    temp=s.copy()
    count=temp.value_counts()
    count.sort_values(ascending=False)
    index=count.index
    lenNotMissing=s.isnull().sum()
    for i in index:
        dict[i]=count[i]/(len(s)-lenNotMissing)
    return dict
cat_col_info_df = cat_col_info_df.agg([missing_ratio, num_values,value_ratios])
display(cat_col_info_df)
```

According to the table above, columns discount_price__currency and price_detail__currency are missing.
