## Data Exploring

### How many rows and how many columns?

### What is the current data type of each column? Are there columns having inappropriate data types?

```python
type_df = pd.DataFrame(udemy_df.dtypes, columns=['Type']).T
type_df
```

- Columns: `published_time`, `last_update_date` are used to represent time, it should be of type `datetime` instead of `object`

- Convert `published_time`, `last_update_date` to datetime

```python
udemy_df['published_time'] = pd.to_datetime(udemy_df['published_time'], format='%Y-%m-%dT%H:%M:%SZ')
udemy_df['last_update_date'] = pd.to_datetime(udemy_df['last_update_date'], format='%Y-%m-%d')
```

### With each numerical column, how are values distributed? 

#### What is the percentage of missing values?

```python
numerical_df = udemy_df.select_dtypes(include='number')
missing_percentage = numerical_df.isnull().mean() * 100
pd.DataFrame(missing_percentage, columns=['Percentage of missing values(%)'])
```

#### Min? max? Are they abnormal?

```python
min_max = numerical_df.agg(['min', 'max'])
min_max
```

- There is nothing abnormal about calculating the minimum and maximum values of numeric columns.

### With each categorical column, how are values distributed?

```python
cat_col_info_df = udemy_df.select_dtypes(include=['object','bool'])
def missing_ratio(s):
    temp=s.isnull().sum()
    result=float(temp)/len(s.index) *100
    return result
def num_values(s):
    temp=s.copy()
    a=list(set(temp))
    has_missing_values = temp.isnull()
    if True in has_missing_values.values:
        return len(a)-1            
    return len(a)
def value_ratios(s):
    dict={}
    temp=s.copy()
    count=temp.value_counts()
    count.sort_values(ascending=False)
    index=count.index
    lenNotMissing=s.isnull().sum()
    for i in index:
        dict[i]= round(count[i]/(len(s)-lenNotMissing), 2)
    return dict
cat_col_info_df = cat_col_info_df.agg([missing_ratio, num_values,value_ratios])
display(cat_col_info_df)
```
